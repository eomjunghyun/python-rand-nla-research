# Section 8.1 Statisticians Citation Accuracy Experiment

논문 Table 2(d)의 Statisticians citation network 실험을 실행한다. 이 데이터셋은 true label이 없으므로 F1/NMI/ARI는 non-random spectral clustering 결과를 기준으로 계산한다.

```python
from pathlib import Path
import json
import sys

import pandas as pd

ROOT = Path.cwd().resolve()
while not (ROOT / "src").exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

from experiments.reference_1_section8_1.exp8_1_remaining_real_networks_live import (
    Exp81RemainingConfig,
    build_table2_like,
    load_dataset,
    run_one_dataset,
    summarize,
    write_table2_markdown,
)
```

```python
DATASET = "statisticians_citation"
REPS = 20

cfg = Exp81RemainingConfig(
    data_dir=ROOT / "data/reference_1_section8_1/processed",
    datasets=(DATASET,),
    reps=REPS,
    seed=2026,
    q=2,
    r=10,
    p_values=(0.7, 0.8),
    outdir=ROOT / "experiments/reference_1_section8_1/results/exp8_1_statisticians_citation_table2_like",
    no_progress=False,
)
cfg.outdir.mkdir(parents=True, exist_ok=True)

ds = load_dataset(cfg.data_dir, DATASET)
ds.meta
```

```python
df_raw, df_pair = run_one_dataset(cfg, ds)
df_summary = summarize(df_raw)
table2_like = build_table2_like(df_summary, cfg.p_values)

raw_csv = cfg.outdir / "statisticians_citation_raw_per_rep.csv"
summary_csv = cfg.outdir / "statisticians_citation_summary_mean_std.csv"
table_csv = cfg.outdir / "statisticians_citation_table2d_like.csv"
table_md = cfg.outdir / "statisticians_citation_table2d_like.md"
pair_csv = cfg.outdir / "statisticians_citation_pairwise_ari_raw.csv"
meta_json = cfg.outdir / "statisticians_citation_meta.json"

df_raw.to_csv(raw_csv, index=False)
df_summary.to_csv(summary_csv, index=False)
table2_like.to_csv(table_csv, index=False)
df_pair.to_csv(pair_csv, index=False)
write_table2_markdown(table2_like, table_md, cfg.reps)
meta_json.write_text(json.dumps({"config": {"dataset": DATASET, "reps": cfg.reps, "seed": cfg.seed, "q": cfg.q, "r": cfg.r, "p_values": list(cfg.p_values), "data_dir": str(cfg.data_dir), "outdir": str(cfg.outdir)}, "dataset_meta": ds.meta}, indent=2), encoding="utf-8")

table2_like
```
