# Section 8.1 Political Blog Accuracy Experiment, Rank 5

Political blog network에서 최종 군집 수는 논문과 같이 2개로 유지하고, spectral embedding rank만 5로 늘려 실험한다.

```python
from pathlib import Path
import json
import sys

ROOT = Path.cwd().resolve()
while not (ROOT / "src").exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

from experiments.reference_1_section8_1.exp8_1_remaining_real_networks_live import (
    Exp81RemainingConfig,
    build_table2_like,
    load_dataset,
    run_one_dataset,
    summarize,
    write_table2_markdown,
)
```

```python
DATASET = "political_blog"
REPS = 20
EMBEDDING_RANK = 5

cfg = Exp81RemainingConfig(
    data_dir=ROOT / "data/reference_1_section8_1/processed",
    datasets=(DATASET,),
    reps=REPS,
    seed=2026,
    q=2,
    r=10,
    p_values=(0.7, 0.8),
    embedding_rank=EMBEDDING_RANK,
    outdir=ROOT / "experiments/reference_1_section8_1/results/exp8_1_political_blog_rank5_table2_like",
    no_progress=False,
)
cfg.outdir.mkdir(parents=True, exist_ok=True)

ds = load_dataset(cfg.data_dir, DATASET)
ds.meta
```

```python
df_raw, df_pair = run_one_dataset(cfg, ds)
df_summary = summarize(df_raw)
table2_like = build_table2_like(df_summary, cfg.p_values)

raw_csv = cfg.outdir / "political_blog_rank5_raw_per_rep.csv"
summary_csv = cfg.outdir / "political_blog_rank5_summary_mean_std.csv"
table_csv = cfg.outdir / "political_blog_rank5_table2b_like.csv"
table_md = cfg.outdir / "political_blog_rank5_table2b_like.md"
pair_csv = cfg.outdir / "political_blog_rank5_pairwise_ari_raw.csv"
meta_json = cfg.outdir / "political_blog_rank5_meta.json"

df_raw.to_csv(raw_csv, index=False)
df_summary.to_csv(summary_csv, index=False)
table2_like.to_csv(table_csv, index=False)
df_pair.to_csv(pair_csv, index=False)
write_table2_markdown(table2_like, table_md, cfg.reps)
meta_json.write_text(json.dumps({"config": {"dataset": DATASET, "reps": cfg.reps, "seed": cfg.seed, "q": cfg.q, "r": cfg.r, "p_values": list(cfg.p_values), "embedding_rank": cfg.embedding_rank, "cluster_count": ds.target_rank, "data_dir": str(cfg.data_dir), "outdir": str(cfg.outdir)}, "dataset_meta": ds.meta}, indent=2), encoding="utf-8")

table2_like
```
